```python
import pandas as pd
btc = pd.read_csv("btcsales.csv")
btc.head()
btc.info()
```

```python
import pandas as pd

# Convert timestamp
btc["Date"] = pd.to_datetime(btc["Timestamp"], unit="s")

# Set Date as index
btc.set_index("Date", inplace=True)

btc.head()
```

```python
# Resample to daily OHLC data
btc_daily = btc.resample("D").agg({
    "Open": "first",
    "High": "max",
    "Low": "min",
    "Close": "last",
    "Volume": "sum"
})

btc_daily.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
btc_daily["Close"].plot(title="BTC Closing Price (Daily)", ylabel="Price (USD)")
plt.show()
```

```python
btc_daily["MA30"] = btc_daily["Close"].rolling(30).mean()
btc_daily["MA90"] = btc_daily["Close"].rolling(90).mean()

btc_daily[["Close","MA30","MA90"]].plot(figsize=(12,6), title="BTC Price with 30 & 90-Day Moving Averages")
plt.show()
```

```python
btc_daily["Daily Return"] = btc_daily["Close"].pct_change()

# Plot distribution of returns
btc_daily["Daily Return"].hist(bins=100, figsize=(10,5))
plt.title("Distribution of BTC Daily Returns")
plt.show()

# Rolling volatility (30-day std deviation of returns)
btc_daily["Volatility30"] = btc_daily["Daily Return"].rolling(30).std()

btc_daily["Volatility30"].plot(figsize=(12,6), title="30-Day Rolling Volatility of BTC")
plt.show()
```
