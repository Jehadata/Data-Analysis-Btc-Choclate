```python
product_sales = df.groupby('Product')['Amount'].sum()

plt.figure(figsize=(8, 8))
plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Product')
plt.show()
```

```python
country_sales = df.groupby('Country')['Amount'].sum()

plt.figure(figsize=(8, 8))
plt.pie(country_sales, labels=country_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Country')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Amount', 'Boxes Shipped']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Sales Amount and Boxes Shipped')
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

# Dropdown for Country selection
country_dropdown = widgets.Dropdown(
    options=df['Country'].unique(),
    description='Country:',
    value=df['Country'].unique()[0]
)

def update_dashboard(country):
    filtered_df = df[df['Country'] == country]
    # Recreate your charts here using filtered_df
    plt.figure(figsize=(10, 6))
    sns.barplot(data=filtered_df.groupby('Sales Person')['Amount'].sum().reset_index().sort_values('Amount', ascending=False),
                x='Sales Person', y='Amount')
    plt.title(f'Total Sales by Sales Person in {country}')
    plt.xlabel('Sales Person')
    plt.ylabel('Sales Amount')
    plt.xticks(rotation=45)
    plt.show()

widgets.interact(update_dashboard, country=country_dropdown)
```
